```python
label_encoder = LabelEncoder()
```

```python
df['Education'] = label_encoder.fit_transform(df['Education'])
df['Marital_Status'] = label_encoder.fit_transform(df['Marital_Status'])
df['Country'] = label_encoder.fit_transform(df['Country'])
```

```python
df
```

```python
 X_marketing = df.iloc[:, [2, 4,28]].values
 X_marketing
```

```python
scaler_marketing = StandardScaler()
X_marketing = scaler_marketing.fit_transform(X_marketing)
```

```python
X_marketing
```

```python
wcss = []
for i in range(1, 11):
  #print(i)
  kmeans_marketing = KMeans(n_clusters=i, random_state=0)
  kmeans_marketing.fit(X_marketing)
  wcss.append(kmeans_marketing.inertia_)
```

```python
wcss
```

```python
grafico = px.line(x = range(1, 11), y = wcss)
grafico.show()
```

```python
df
```

```python
kmeans_marketing = KMeans(n_clusters=4, random_state=0)
rotulos = kmeans_marketing.fit_predict(X_marketing)
```

```python
spectral = SpectralClustering(n_clusters=4, random_state=0)
rotulos_spectral = spectral.fit_predict(X_marketing)
```

```python
affinity_propagation = AffinityPropagation(random_state=0)
rotulos_affinity = affinity_propagation.fit_predict(X_marketing)
```

```python
gmm = GaussianMixture(n_components=4, random_state=0)
rotulos_gmm = gmm.fit_predict(X_marketing)
```

```python
hdbscan_clusterer = hdbscan.HDBSCAN(min_cluster_size=10, gen_min_span_tree=True)
rotulos_hdbscan = hdbscan_clusterer.fit_predict(X_marketing)
```
