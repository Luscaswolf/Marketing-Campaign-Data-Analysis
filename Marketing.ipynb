```python
   df['Total_Purchases'] = df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds'] + df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
df.Education = df.Education.str.replace(' ', '')
```

```python
len(np.unique(df['Education']))
```

```python
np.unique(df['Education'])
```

```python
label_encoder = LabelEncoder()
```

```python
df['Education'] = label_encoder.fit_transform(df['Education'])
df['Marital_Status'] = label_encoder.fit_transform(df['Marital_Status'])
df['Country'] = label_encoder.fit_transform(df['Country'])
```

```python
df
```

```python
 X_marketing = df.iloc[:, [2, 4,28]].values
 X_marketing
```

```python
scaler_marketing = StandardScaler()
X_marketing = scaler_marketing.fit_transform(X_marketing)
```

```python
X_marketing
```

```python
wcss = []
for i in range(1, 11):
  #print(i)
  kmeans_marketing = KMeans(n_clusters=i, random_state=0)
  kmeans_marketing.fit(X_marketing)
  wcss.append(kmeans_marketing.inertia_)
```

```python
wcss
```

```python
grafico = px.line(x = range(1, 11), y = wcss)
grafico.show()
```

```python
df
```

```python
kmeans_marketing = KMeans(n_clusters=4, random_state=0)
rotulos = kmeans_marketing.fit_predict(X_marketing)
```

```python
grafico = px.scatter(x=X_marketing[:, 0], y=X_marketing[:, 1], color=rotulos)
grafico.show()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X_marketing)
```

```python
df_tsne = pd.DataFrame(X_tsne, columns=['Dim1', 'Dim2'])
df_tsne['Cluster'] = rotulos
```

```python
grafico_tsne = px.scatter(df_tsne, x='Dim1', y='Dim2', color='Cluster', title='Clusters de Clientes com t-SNE', labels={'Dim1': 'DimensÃ£o 1', 'Dim2': 'DimensÃ£o 2'})
grafico_tsne.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)  # Reduzir para 2 dimensÃµes para visualizaÃ§Ã£o
X_pca = pca.fit_transform(X_marketing)
```

```python
df_pca = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
```

```python
kmeans_marketing = KMeans(n_clusters=2, random_state=0)
rotulos = kmeans_marketing.fit_predict(X_pca)
df_pca['Cluster'] = rotulos
```

```python
grafico_pca = px.scatter(df_pca, x='PCA1', y='PCA2', color='Cluster',
                         title='Clusters de Clientes com PCA',
                         labels={'PCA1': 'Componente Principal 1', 'PCA2': 'Componente Principal 2'})
grafico_pca.show()
```
